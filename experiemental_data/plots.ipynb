```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
rewards_noPR = 'prioritized_replay/info_rewards-no_prior_replay.csv'
rewards_PR = 'prioritized_replay/info_rewards-prior_replay.csv'
noPR = []

with open(rewards_noPR, mode='r') as csv_file:
    data_iter = csv.reader(csv_file, delimiter=',')
    data = [data for data in data_iter]
    noPR = np.array(data)
    
with open(rewards_PR, mode='r') as csv_file:
    data_iter = csv.reader(csv_file, delimiter=',')
    data = [data for data in data_iter]
    PR = np.array(data)
```

```python
noPR_wall_time=noPR[1:,0]
noPR_step=noPR[1:,1]
noPR_value=noPR[1:,2]

PR_wall_time=PR[1:,0]
PR_step=PR[1:,1]
PR_value=PR[1:,2]
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, label='No prioritized replay')
ax2=fig.add_subplot(111, label="2", frame_on=False)

ax.plot(noPR_step, noPR_value, color="blue", label='No prioritized replay')
# ax.set_xlabel("Time step", color="blue")
# ax.set_ylabel("No prioritized replay", color="blue")
# ax.tick_params(axis='x', colors="C0")
# ax.tick_params(axis='y', colors="C0")

# ax2 = ax.twiny()
ax2.plot(PR_step, PR_value, color="red", label='Prioritized replay')
# ax2.xaxis.tick_top()
# ax2.yaxis.tick_right()
# ax2.set_xlabel('Time step', color="red") 
# ax2.set_ylabel('Prioritized replay', color="red")       
# ax2.xaxis.set_label_position('top') 
# ax2.yaxis.set_label_position('right') 
# ax2.tick_params(axis='x', colors="C1")
# ax2.tick_params(axis='y', colors="C1")

mini = min(min(noPR_value), min(PR_value))
maxi = max(max(noPR_value), max(PR_value))
plt.yticks(np.arange(float(mini), float(maxi), 0.1))

plt.show()

# plt.plot(noPR_step, noPR_value, color='blue', label='No prioritized replay')
# plt.plot(PR_step, PR_value, color='red', label='Prioritized replay')
# plt.show()
```
